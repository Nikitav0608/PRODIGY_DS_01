```python
# Define the color palette
palette_color = sns.color_palette('deep')

# Calculate value counts
df = data_Pokemon['Type 1'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 8))  # Adjust the figure size as needed
plt.pie(df, labels=df.index, autopct='%.1f%%', labeldistance=0.9, colors=palette_color)
plt.title('Distribution of Pokemon Types')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
data_Pokemon['Legendary'].value_counts().plot(kind='bar',color='lightgreen',title='Legendary Check')
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data = data_Pokemon,hue='Generation' , x='Generation',palette='pastel',legend='auto')
plt.title('Number of Pokemons Gen-Wise')
plt.show()
```

Histogram Plot

```python
sns.histplot(data =data_Pokemon['Total'],color='salmon')
plt.xlabel=('Total Powers of the Pokemons')
plt.show()
```

Extracting Top 10 powerful Pokemons

```python
top_10 = data_Pokemon.sort_values('Total',ascending=False).head(10)
```

```python
plt.figure(figsize=(15, 6))
plt.subplot(2,2,1)
sns.barplot(data=top_10,x='Total',y='Name',palette='plasma')
plt.title('10 Most powerful Pokemons')
plt.show()
```

Generation Wise the Pokemons that have Maximum Total Power

```python
gen=data_Pokemon.groupby('Generation')
```

```python
top_in_gen= gen.max('Total')
top_in_gen
```

```python
top_in_gen.reset_index()
```

```python

sns.barplot(data=top_in_gen,x='Generation',y='Total',palette='crest')
plt.title('Gen-Wise MaxPower')
plt.show()
```

Generation Wise the Pokemons that have Minimum Total Power

```python
top_in_gen= gen.min('Total')
top_in_gen
```

```python
top_in_gen.reset_index()
```

```python

sns.barplot(data=top_in_gen,x='Generation',y='Total',palette='deep')
plt.title('Gen-Wise MinPower')
plt.show()
```
